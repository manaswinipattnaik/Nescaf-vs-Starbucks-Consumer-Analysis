## Visualizations
Bar Chart - Age Distribution

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set_theme(style="whitegrid")

# Filter data for Age brackets
age_data = df1[df1["Population Characteristics"] == "Age (Brackets)"]

# Create bar chart
plt.figure(figsize=(12,6))
sns.barplot(x="Element", y="Index 1", data=age_data, color="blue", label="Nescafe")
sns.barplot(x="Element", y="Index2", data=age_data, color="orange", label="Starbucks")
plt.title("Age Group Index Comparison")
plt.axhline(100, color="black", linestyle="--")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

### Histogram - Income Distribution

```python
# Select numeric columns
numeric_cols = df1.select_dtypes(include=['number']).columns

# Create histograms for numeric attributes
fig, axes = plt.subplots(len(numeric_cols), 1, figsize=(8, len(numeric_cols) * 4))

for i, col in enumerate(numeric_cols):
    sns.histplot(df1[col], kde=True, ax=axes[i])
    axes[i].set_title(f"Distribution of {col}")
    axes[i].set_xlabel(col)

plt.tight_layout()
plt.show()
```

### Identify characteristics where Nescafe or Starbucks over-indexes:
Over (Above 110)
Under (Below 90)
Neutral (Between 90-110)

```python
# Convert Index columns to numeric (if needed)
df1["Index 1"] = pd.to_numeric(df1["Index 1"], errors="coerce")
df1["Index2"] = pd.to_numeric(df1["Index2"], errors="coerce")

# Define function to classify indexing
def classify_index(value):
    if value > 110:
        return "Over"
    elif value < 90:
        return "Under"
    else:
        return "Neutral"

# Apply classification
df1["Significant_Nescafe"] = df1["Index 1"].apply(classify_index)
df1["Significant_Starbucks"] = df1["Index2"].apply(classify_index)

# Display first few rows to verify
df1[["Element", "Index 1", "Significant_Nescafe", "Index2", "Significant_Starbucks"]].head() 
```

### Top over-indexed groups for Nescafe

```python
top_nescafe = df1[df1["Significant_Nescafe"] == "Over"].sort_values("Index 1", ascending=False).head(5)
print("Top Over-Indexed Groups for Nescafe:")
print(top_nescafe[["Element", "% of Nescafe Shoppers", "Index 1"]])
```

### Top over-indexed groups for Starbucks

```python
top_starbucks = df1[df1["Significant_Starbucks"] == "Over"].sort_values("Index2", ascending=False).head(5)
print("\nTop Over-Indexed Groups for Starbucks:")
print(top_starbucks[["Element", "% of Starbucks Shoppers", "Index2"]])
```
